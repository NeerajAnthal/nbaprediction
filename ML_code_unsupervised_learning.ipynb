```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Load your dataset (replace 'your_data.csv' with your actual file path)
df = pd.read_csv('all_seasons.csv')

```

```python

df['draft_round'] = df['draft_round'].replace('Undrafted', np.nan)
df['draft_number'] = df['draft_number'].replace('Undrafted', np.nan)

```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import pandas as pd

# Select features related to player performance
features = df[['pts', 'reb', 'ast', 'net_rating', 'usg_pct', 'ts_pct']]

# Step 1: Standardize all features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
scaled_df = pd.DataFrame(scaled_features, columns=features.columns)

# Step 2: Apply Min-Max scaling to all features with range -1 to 1
min_max_scaler = MinMaxScaler(feature_range=(-1, 1))
all_features_scaled = min_max_scaler.fit_transform(scaled_df)

# Convert back to DataFrame
scaled_df = pd.DataFrame(all_features_scaled, columns=features.columns)

scaled_df.head()
```

```python
# Step 2: Apply K-Means Clustering
kmeans = KMeans(n_clusters=4, random_state=42)
df['cluster'] = kmeans.fit_predict(scaled_features)

# Step 3: Assign labels to clusters based on relative performance
# Analyze the centroids and assign labels
centroids = kmeans.cluster_centers_
```

```python



# Labels based on performance levels (adjust if necessary after examining centroids)
cluster_labels = ['very good', 'good', 'average', 'bad']  # These can be changed after viewing cluster distribution
cluster_order = centroids[:, 0].argsort()  # Sort clusters by primary feature (e.g., 'pts')
labels = [cluster_labels[i] for i in cluster_order]

```

```python
# Map each cluster to its respective label
df['performance_label'] = df['cluster'].map(lambda x: labels[x])

# Output: Show dataset with the new column 'performance_label'
print(df[['player_name', 'pts', 'reb', 'ast', 'net_rating', 'usg_pct', 'ts_pct', 'performance_label']].head())
```
